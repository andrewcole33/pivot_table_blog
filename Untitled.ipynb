```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt


plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.Year = df.Year.astype('str')
df.Year = [x.split('.')[0] for x in df.Year]
```

# Grouping Data Using Pivot Table Indices

```python
table = pd.pivot_table(data = df, index = 'Year', values = ['EU_Sales', 'JP_Sales', 'NA_Sales', 'Global_Sales', 'Other_Sales'])
table.drop(table.tail(1).index, inplace = True)
table.head(10)
```

```python
t = table.plot(kind = 'line', figsize = (20, 10), lw = 2.5)
t.set_xlabel('Year', fontsize = 20)
t.set_ylabel('Sales (Millions)', fontsize = 20)
t.set_title('Sales by Year - Trends', fontsize = 20)
t.legend(fontsize = 20, edgecolor = 'black', facecolor = 'white', fancybox = True)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.tight_layout()
```

```python
table2 = pd.pivot_table(data = df, index = 'Genre', values = ['EU_Sales', 'JP_Sales', 'NA_Sales', 'Global_Sales', 'Other_Sales'])
table2
```

```python
x = table2.plot(kind = 'bar', figsize = (20,10))
x.set_xlabel(xlabel = 'Genre', size = 20)
x.set_ylabel(ylabel = 'Sales (Millions)', size = 20)
plt.xticks(rotation = 45, fontstyle = 'normal', fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15, facecolor = 'white', edgecolor = 'black', fancybox = True)
plt.tight_layout()
```

# Multi-Index Pivot Table

```python
table3 = pd.pivot_table(df, index = ['Publisher', 'Genre'], values = ['EU_Sales', 'JP_Sales', 'NA_Sales'])
table3
```

# Pivots - Different Aggregation Techniques

```python
table4 = pd.pivot_table(df, index = ['Publisher', 'Genre'], aggfunc = {'EU_Sales':np.mean, 'Global_Sales':np.sum, 'JP_Sales':np.median})
table4.columns = ['EU_Sales_mean', 'Global_Sales_sum', 'JP_Sales_median']
table4
```

# Showing Relationships w/ Pivot Tables

```python
table5 = pd.pivot_table(df, index = ['Year'], columns = ['Genre'], values = ['Global_Sales'], aggfunc = np.sum, fill_value = 0)
table5
```
